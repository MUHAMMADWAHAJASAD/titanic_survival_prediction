```python
# Allow imports from src folder
import sys
sys.path.append("../src")

# Import functions
from data_preprocessing import load_data, clean_data
from feature_engineering import extract_title,impute_age_by_group,create_family_features,create_is_alone

import pandas as pd
import os

# Ensure processed folder exists
os.makedirs("../data/processed", exist_ok=True)

# 1. Load raw data
df = load_data("Titanic-Dataset.csv")
print("Raw Data Shape:", df.shape)
df.head()
```

```python
clean_df = clean_data(df)
print("After Cleaning Shape:", clean_df.shape)
clean_df.head()
```

```python
fe_df = extract_title(clean_df)
```

```python
fe_df = impute_age_by_group(fe_df)
```

```python
fe_df = create_family_features(fe_df)
```

```python
fe_df = create_is_alone(fe_df)
```

```python
print("After Feature Engineering Shape:", fe_df.shape)
fe_df.head()
```

```python
processed_path = "../data/processed/cleaned_titanic.csv"
fe_df.to_csv(processed_path, index=False)
print(f"Processed data saved to: {processed_path}")
```
